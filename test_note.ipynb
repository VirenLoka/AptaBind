```python
import pandas as pd

ct_df= pd.read_csv('data/CT-20.tsv', sep = '\t')
ct_df = ct_df.iloc[:2000]
ct_df['label'] = 0

bc_df = pd.read_csv('data/BC-6.tsv', sep = '\t')
bc_df = bc_df.iloc[:2000]
bc_df['label'] = 0

dk_df = pd.read_csv('data/DK-30.tsv', sep = '\t')
dk_df = dk_df.iloc[:4000]
dk_df['label'] = 1
print(ct_df.shape, dk_df.shape)
main_df =  pd.concat([dk_df, ct_df,bc_df], ignore_index = True)
main_df.shape
new_df = []
for index, row in main_df.iterrows():
    row['sequence_id'] = f'seq_{index}'
    new_df.append(row)

main_df.to_csv('data/backup_cross_main.tsv',sep = '\t', index = False)

new_df = pd.DataFrame(new_df)
print(new_df.columns)
new_df = new_df[['sequence_id','sequence','label','score']]
new_df.rename(columns = {'score': 'enrichment_score'},inplace = True)
new_df.to_csv('data/cross_main.tsv', sep = '\t', index = False)
```
